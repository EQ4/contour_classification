```python
import glob
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
olap_vals = np.arange(0, 1, 0.1)
```

```python
meltype=1
```

```python
if meltype == 1:
    mel_dir = mel_dir1
else:
    mel_dir = mel_dir2
    
all_cols = ['clf mean', 'clf std', 'melscore mean', 'melscore std',
            'melodiness mean', 'melodiness std']

olap_str = [str(olap) for olap in olap_vals]

clf_dict_avg = {k: {} for k in olap_str}
melscore_dict_avg = {k: {} for k in olap_str}
melodiness_dict_avg = {k: {} for k in olap_str}

clf_dict_std = {k: {} for k in olap_str}
melscore_dict_std = {k: {} for k in olap_str}
melodiness_dict_std = {k: {} for k in olap_str}

for i, olap in enumerate(olap_vals):
    olap_dir = "olap_%s" % olap

    clf_dat = []
    melscore_dat = []
    melodiness_dat = []
    for split in range(1, 6):
        split_dir = "splitnum_%s" % split
        temp_path = os.path.join(mel_dir, split_dir, olap_dir)
        
        clf_score_fpath = os.path.join(temp_path, "classifier_scores.csv")
        temp_df = pd.read_csv(clf_score_fpath, index_col=0)
        
        x = np.fromstring(temp_df.loc['f1', 'test'].strip('[').strip(']'), sep=' ', dtype='float')
        temp_df.loc['f1 nonmel', 'test'] = x[0]
        temp_df.loc['f1 mel', 'test'] = x[1]
        temp_df.loc['f1 mean', 'test'] = np.mean(x)
        clf_cols = ['accuracy', 'auc score', 'f1 nonmel', 'f1 mel', 'f1 mean']
        clf_dat.append(temp_df.loc[clf_cols, :].convert_objects(convert_numeric=True))
        
        mel_score_fpath = os.path.join(temp_path, "validate_mel_score_summary.csv")
        temp_df = pd.read_csv(mel_score_fpath, index_col=0)
        melscore_dat.append(temp_df.iloc[1:3, :].convert_objects(convert_numeric=True))
        
        melodiness_fpath = os.path.join(temp_path, "melodiness_scores.csv")
        temp_df = pd.read_csv(melodiness_fpath, index_col=0)
        x = np.fromstring(temp_df.loc['f1', 'test'].strip('[').strip(']'), sep=' ', dtype='float')
        temp_df.loc['f1 nonmel', 'test'] = x[0]
        temp_df.loc['f1 mel', 'test'] = x[1]
        temp_df.loc['f1 mean', 'test'] = np.mean(x)
        melodiness_dat.append(temp_df.loc[clf_cols, :].convert_objects(convert_numeric=True))
    
    clf_dict_avg[str(olap)] = pd.concat(clf_dat).groupby(level=0).mean()
    clf_dict_std[str(olap)] = pd.concat(clf_dat).groupby(level=0).std()

    melscore_dict_avg[str(olap)] = pd.concat(melscore_dat).groupby(level=0).mean()
    melscore_dict_std[str(olap)] = pd.concat(melscore_dat).groupby(level=0).std()
    
    melodiness_dict_avg[str(olap)] = pd.concat(melodiness_dat).groupby(level=0).mean()
    melodiness_dict_std[str(olap)] = pd.concat(melodiness_dat).groupby(level=0).std()

clf_panel_avg = pd.Panel(data=clf_dict_avg)
clf_panel_std = pd.Panel(data=clf_dict_std)

melscore_panel_avg = pd.Panel(data=melscore_dict_avg)
melscore_panel_std = pd.Panel(data=melscore_dict_std)

melodiness_panel_avg = pd.Panel(data=melodiness_dict_avg)
melodiness_panel_std = pd.Panel(data=melodiness_dict_std)

melscore_panel_avg.loc[:, 'mean', :]
```

```python
c1, c2, c3, c4, c5 = sns.color_palette('deep', 5)

xlims = [-0.1, 1.0]
ylims = [0, 1]

fig = plt.figure(figsize=(5,8))
plt.subplot(3,1,1)
p1 = plt.errorbar(olap_vals, melscore_panel_avg.loc[:, 'mean', 'VR'], 
                  melscore_panel_std.loc[:, 'mean', 'VR'], label='Voicing Recall', 
                  linestyle='-', marker='o', color=c1, alpha=1, capsize=0)

p2 = plt.errorbar(olap_vals, melscore_panel_avg.loc[:, 'mean', 'VFA'], 
                  melscore_panel_std.loc[:, 'mean', 'VFA'], label='Voicing False Alarm', 
                  linestyle='-', marker='o', color=c2, alpha=1, capsize=0)
plt.xlim(xlims)
plt.ylim(ylims)
plt.legend(loc=4, fontsize=12)
plt.title("Melody %s Metrics vs. Overlap Threshold" % meltype, fontsize=12)
plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')

plt.subplot(3,1,2)
p3 = plt.errorbar(olap_vals+0.01, melscore_panel_avg.loc[:, 'mean', 'RPA'], 
                  melscore_panel_std.loc[:, 'mean', 'RPA'], label='Raw Pitch Accuracy', 
                  linestyle='-', marker='o', color=c3, alpha=1, capsize=0)
p4 = plt.errorbar(olap_vals-0.01, melscore_panel_avg.loc[:, 'mean', 'RCA'], 
                  melscore_panel_std.loc[:, 'mean', 'RCA'], label='Raw Chroma Accuracy', 
                  linestyle='-', marker='o', color=c4, alpha=1, capsize=0)
plt.xlim(xlims)
plt.ylim(ylims)
plt.legend(loc=4, fontsize=12)
plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')

ax = plt.subplot(3,1,3)
p5 = plt.errorbar(olap_vals, melscore_panel_avg.loc[:, 'mean', 'OA'], 
                  melscore_panel_std.loc[:, 'mean', 'OA'], label='Overall Accuracy', 
                  linestyle='-', marker='o', color=c5, alpha=1, capsize=0)
plt.xlim(xlims)
plt.ylim(ylims)
plt.legend(loc=4, fontsize=12)
plt.xlabel('Overlap Threshold', fontsize=12)

# plt.show()
plt.savefig('/Users/rachelbittner/Desktop/mel%s_metrics_v_olap.eps' % meltype, format='eps')
```
